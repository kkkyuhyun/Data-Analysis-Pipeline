파일 저장 및 열기: DataFrame 객체를 엑셀,csv,txt 등의 파일 형태로 저장하고 열기 

```python
import pandas as pd

# 주어진 데이터 딕셔너리
data = {
    '이름': ['채치수', '정대만', '송태섭', '서태웅', '강백호', '황태산', '윤대형'],
    '학교': ['북산고'] * 5 + ['능남고'] * 2,
    '키': [197, 184, 168, 187, 188, 202, 188],
    '국어': [90, 40, 80, 40, 15, 80, 55],
    '영어': [85, 35, 75, 60, 20, 100, 65],
    '수학': [100, 50, 70, 70, 10, 95, 45],
    '과학': [95, 55, 80, 75, 35, 85, 40],
    '사회': [85, 25, 75, 80, 10, 80, 35],
    'SW특기': ['Python', 'Java', 'Javascript', '', 'C', 'Python', 'C#']
}

# DataFrame 생성
df = pd.DataFrame(data)
df.index = ['1번', '2번', '3번', '4번', '5번', '6번', '7번']

#인덱스
df.index.name='지원번호'

# 생성된 DataFrame 출력
df
```

저장하기 - CSV 파일로 저장

```python
df.to_csv('score.csv',encoding='utf-8-sig')
```

```python
df.to_csv('score.csv',encoding='utf-8-sig',index=False)
```

텍스트(txt파일로)저장

```python
df.to_csv('score txt',sep='\t') #\tab으로 구분
```

엑셀 파일로 저장

```python
df.to_excel('score. xlsx')
```

```python
# Assuming df is already defined and has an index name set as '지원번호'

# Write DataFrame to Excel file without specifying the engine
df.to_excel('score.xlsx', engine='xlsxwriter')
```
