```python
import pandas as pd
import numpy as np

# 데이터 생성
data = {
    '이름': ['채치수', '정대만', '송태섭', '서태웅', '강백호', '황태산', '윤대형'],
    '학교': ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고'],
    '키': [197, 184, 168, 187, 188, 202, 188],
    '국어': [90, 40, 80, 40, 15, 80, 55],
    '영어': [85, 35, 75, 60, 20, 100, 65],
    '수학': [100, 50, 70, 70, 10, 95, 45],
    '과학': [95, 55, 80, 75, 35, 85, 40],
    '사회': [85, 25, 75, 80, 10, 80, 35],
    'SW특기': ['Python', 'Java', 'Javascript', np.nan, 'C', 'Python', 'C#']
}

# DataFrame 생성
df = pd.DataFrame(data)

# NaN 값을 'Unknown'으로 대체
df['SW특기'] = df['SW특기'].fillna('Unknown')

# 'SW특기' 열에서 'Java'를 포함하는 행 필터링
filt = df['SW특기'].str.contains('Java')
filtered_df = df[filt]

# 결과 출력
print(filtered_df)
```
