```python
import matplotlib.pyplot as plt
import matplotlib

# 데이터 생성
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# 그래프 그리기
plt.plot(x, y)
plt.title('꺾은선 그래프')

# 글꼴 및 글꼴 크기 설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15

# 그래프 표시
plt.show()
```

```python
import matplotlib.font_manager as fm

fonts = fm.fontManager.ttflist
for font in fonts:
    print(font.name)
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프')
```

```python
import matplotlib.pyplot as plt

plt.plot([-1, 0, 1], [-5, -1, 2])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.grid(True)
plt.show()
```
