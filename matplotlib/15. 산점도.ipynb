```python
import matplotlib.pyplot as plt


import pandas as pd
df = pd.DataFrame({
    '지원번호': ['1번', '2번', '3번', '4번', '5번', '6번', '7번'],
    '이름': ['채치수', '정대만', '송태섭', '서태웅', '강백호', '황태산', '윤대형'],
    '학교': ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고'],
    '키': [197, 184, 168, 187, 188, 202, 188],
    '국어': [90, 40, 80, 40, 15, 80, 55],
    '영어': [85, 35, 75, 60, 20, 100, 65],
    '수학': [100, 50, 70, 70, 10, 95, 45],
    '과학': [95, 55, 80, 75, 35, 85, 40],
    '사회': [85, 25, 75, 80, 10, 80, 35],
    'SW특기': ['Python', 'Java', 'Javascript', None, 'C', 'Python', 'C#'],
    '학년': [3, 3, 2, 1, 1, 3, 2]
})

sizes = df['학년'] * 100

plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.3)
plt.xlabel('영어 점수')
plt.ylabel('수학 점수')
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.5,orientation='horizontal')  
plt.show()
```
