```python
import matplotlib.pyplot as plt

labels = ['강백호', '서태웅', '정대만']
values = [180, 181, 183]

bars = plt.bar(labels, values)
bars[0].set_hatch('/')
bars[1].set_hatch('x')
bars[2].set_hatch('_') #hatch스타일 활용하여 사용 가능

plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['강백호', '서태웅', '정대만']
values = [180, 181, 183]

bars = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bars):
    plt.text(rect.get_x() + rect.get_width() / 2.0, rect.get_height(), values[idx], ha='center', va='bottom')

plt.show()
```
